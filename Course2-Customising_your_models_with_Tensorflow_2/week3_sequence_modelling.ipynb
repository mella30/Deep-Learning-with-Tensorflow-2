## Padding and Masking Sequence Data

```python
# Load the imdb data set

(x_train, y_train), (x_test, y_test) = imdb.load_data()
```

#### Preprocess the data with padding

```python
# Inspect the input data shape

x_train.shape  # resulting arrays are ragged (not all of the same length)
```

```python
# Pad the inputs to the maximum length using maxlen so that the array is not ragged anymore
# sequences longer than maxlen will be truncated, sequences shorter than that will be padded

padded_x_train = sequence.pad_sequences(x_train, maxlen=300, padding='post', truncating='pre')
```

```python
# Inspect the output data shape

padded_x_train.shape
```

#### Create a Masking layer

```python
# Import numpy 

import numpy as np
```

```python
# Masking expects to see (batch, sequence, features)
# Create a dummy feature dimension using expand_dims

padded_x_train = np.expand_dims(padded_x_train, axis=-1)
```

```python
# Create a Masking layer 

tf_x_train = tf.convert_to_tensor(padded_x_train, dtype='float32')
masking_layer = Masking(mask_value=0.0)
```

```python
# Pass tf_x_train to it

masked_x_train = masking_layer(tf_x_train)
```

```python
# Look at the dataset

tf_x_train
masked_x_train
# are both the same (added an additional property with mask)
```

#### Load and preprocess the IMDb data

```python
# A function to load and preprocess the IMDB dataset

def get_and_pad_imdb_dataset(num_words=10000, maxlen=None, index_from=2):
    from tensorflow.keras.datasets import imdb

    # Load the reviews
    (x_train, y_train), (x_test, y_test) = imdb.load_data(path='imdb.npz',
                                                          num_words=num_words,
                                                          skip_top=0,
                                                          maxlen=maxlen,
                                                          start_char=1,
                                                          oov_char=2,
                                                          index_from=index_from)

    x_train = sequence.pad_sequences(x_train, 
                                     maxlen=None,
                                     padding='pre',
                                     truncating='pre',
                                     value=0)
    
    x_test = sequence.pad_sequences(x_test,
                                    maxlen=None,
                                    padding='pre',
                                    truncating='pre',
                                    value=0)
    return (x_train, y_train), (x_test, y_test)
```

```python
# Load the dataset

(x_train, y_train), (x_test, y_test) = get_and_pad_imdb_dataset()
```

```python
# A function to get the dataset word index

def get_imdb_word_index(num_words=10000, index_from=2):
    imdb_word_index = tf.keras.datasets.imdb.get_word_index(
                                        path='imdb_word_index.json')
    imdb_word_index = {key: value + index_from for
                       key, value in imdb_word_index.items() if value <= num_words-index_from}
    return imdb_word_index
```

```python
# Get the word index

imdb_word_index = get_imdb_word_index()  # encode a sentensce
```

```python
# Swap the keys and values of the word index

inv_imdb_word_index = {value: key for key, value in imdb_word_index.items()}  # decode a sentence
```

```python
# View the first dataset example sentence

[inv_imdb_word_index[index] for index in x_train[100] if index > 2]
```

#### Build an Embedding layer into a model

```python
# Get the maximum token value

max_index_value = max(imdb_word_index.values())
```

```python
# Specify an embedding dimension

embedding_dim = 16
```

```python
# Build a model using Sequential:
#     1. Embedding layer
#     2. GlobalAveragePooling1D
#     3. Dense

model = tf.keras.Sequential([
                             Embedding(input_dim=max_index_value+1, output_dim=embedding_dim, mask_zero=False),
                             GlobalAveragePooling1D(),
                             Dense(1, activation='sigmoid')

])
```

```python
# Functional API refresher: use the Model to build the same model

review_sequence = Input((None, ))
embedding_sequence = Embedding(input_dim=max_index_value+1, output_dim=embedding_dim)(review_sequence)
average_embedding = GlobalAveragePooling1D()(embedding_sequence)
positive_probability = Dense(1, activation='sigmoid')(average_embedding)

model = tf.keras.Model(inputs=review_sequence, outputs=positive_probability)
```

```python
model.summary()
```

#### Compile, train, and evaluate the model

```python
# Compile the model with a binary cross-entropy loss

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model using .fit(), savng its history

history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test), validation_steps=20)
```

```python
# Plot the training and validation accuracy

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

history_dict = history.history

acc      = history_dict['accuracy']
val_acc  = history_dict['val_accuracy']
loss     = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

plt.figure()
plt.plot(epochs, acc, marker='.', label='Training acc')
plt.plot(epochs, val_acc, marker='.', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Classification accuracy')
plt.legend(loc='lower right')
plt.ylim(0, 1);
```

#### Load and transform the IMDB review sentiment dataset

```python
# A function to load and preprocess the IMDB dataset

def get_and_pad_imdb_dataset(num_words=10000, maxlen=None, index_from=2):
    from tensorflow.keras.datasets import imdb

    # Load the reviews
    (x_train, y_train), (x_test, y_test) = imdb.load_data(path='imdb.npz',
                                                          num_words=num_words,
                                                          skip_top=0,
                                                          maxlen=maxlen,
                                                          start_char=1,
                                                          oov_char=2,
                                                          index_from=index_from)

    x_train = sequence.pad_sequences(x_train,
                                     maxlen=None,
                                     padding='pre',
                                     truncating='pre',
                                     value=0)
    
    x_test = sequence.pad_sequences(x_test,
                                    maxlen=None,
                                    padding='pre',
                                    truncating='pre',
                                    value=0)
    return (x_train, y_train), (x_test, y_test)
```

## Stacked RNNs and the Bidirectional wrapper

#### Load and transform the IMDb review sentiment dataset

```python
# A function to load and preprocess the IMDB dataset

def get_and_pad_imdb_dataset(num_words=10000, maxlen=None, index_from=2):
    from tensorflow.keras.datasets import imdb

    # Load the reviews
    (x_train, y_train), (x_test, y_test) = imdb.load_data(path='imdb.npz',
                                                          num_words=num_words,
                                                          skip_top=0,
                                                          maxlen=maxlen,
                                                          start_char=1,
                                                          oov_char=2,
                                                          index_from=index_from)

    x_train = sequence.pad_sequences(x_train,
                                     maxlen=None,
                                     padding='pre',
                                     truncating='pre',
                                     value=0)
    
    x_test = sequence.pad_sequences(x_test,
                                    maxlen=None,
                                    padding='pre',
                                    truncating='pre',
                                    value=0)
    return (x_train, y_train), (x_test, y_test)
```

```python
# Load the dataset

(x_train, y_train), (x_test, y_test) = get_and_pad_imdb_dataset(maxlen=250, num_words=5000)
```

```python
# A function to get the dataset word index

def get_imdb_word_index(num_words=10000, index_from=2):
    imdb_word_index = tf.keras.datasets.imdb.get_word_index(
                                        path='imdb_word_index.json')
    imdb_word_index = {key: value + index_from for
                       key, value in imdb_word_index.items() if value <= num_words-index_from}
    return imdb_word_index
```

```python
# Get the word index using get_imdb_word_index()

imdb_word_index = get_imdb_word_index(num_words=5000)
```

#### Build stacked and bidirectional recurrent models

```python
# Get the maximum index value and specify an embedding dimension

max_index_value = max(imdb_word_index.values())
embedding_dim = 16
```

```python
# Using Sequential, build a stacked LSTM model via return_sequences=True

model = tf.keras.models.Sequential([
                                    Embedding(input_dim=max_index_value+1, output_dim=embedding_dim, mask_zero=True),
                                    LSTM(32, return_sequences=True),  # full sequence returned
                                    LSTM(32, return_sequences=False),   # only last state returned
                                    Dense(1, activation='sigmoid')
])
```

```python
# Using Sequential, build a bidirectional RNN with merge_mode='sum' (to pass a sequence in forward and backward direction)

# keywords:
# layer: forward layer (both directions if only given one argument)
# merge mode: how layers are combined (can be sum, mul, concat, ave)
# backward layer: backward layer (needs go backwards set to True)

model = tf.keras.models.Sequential([
                                   Embedding(input_dim=max_index_value+1, output_dim=embedding_dim, mask_zero=True),
                                   Bidirectional(layer=LSTM(8),  merge_mode='sum',
                                                  backward_layer=GRU(8, go_backwards=True)),
                                   Dense(1, activation='sigmoid')
])
```

```python
# Create a model featuring both stacked recurrent layers and a bidirectional layer

model = tf.keras.models.Sequential([
                                    Embedding(input_dim=max_index_value+1, output_dim=embedding_dim, mask_zero=True),
                                    Bidirectional(layer=LSTM(8, return_sequences=True), merge_mode='concat'),
                                    GRU(8, return_sequences=False),
                                    Dense(1, activation='sigmoid')
])
```

#### Compile and fit the model

```python
# Compile the model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model, saving its history

history = model.fit(x_train, y_train, epochs=3, batch_size=32)
```

```python
# Plot the training and validation accuracy

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

history_dict = history.history

acc      = history_dict['accuracy']
loss     = history_dict['loss']

epochs = range(1, len(acc) + 1)

plt.figure(figsize=(14,5))
plt.plot(epochs, acc, marker='.', label='Training acc')
plt.title('Training accuracy')
plt.xlabel('Epoch')
plt.ylabel('Classification accuracy')
plt.legend(loc='lower right')
plt.ylim(0, 1);
```
